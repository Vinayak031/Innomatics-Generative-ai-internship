## Problem 1: Mobile Recharge Validation System

```python

def check_recharge(amount, plan):

    valid_plans = [199, 299, 399, 599]

    return amount >= 50 and plan in valid_plans


print("Output:", check_recharge(100, 299))
```

## Problem 2: Inventory Reorder Alert Program

```python

def reorder_alert(products):

    for item, stock in products.items():

        if stock < 15:

            print(item, "- Reorder Alert")

        else:

            print(item, "- Stock OK")


print("Output:")
reorder_alert({"Rice": 20, "Oil": 10, "Sugar": 14})
```

## Problem 3: Student Result Processing Program

```python

def student_result(marks):

    total = 0

    for m in marks:

        total += m

    avg = total / len(marks)

    if avg >= 50:

        return avg, "Pass"

    return avg, "Fail"


print("Output:", student_result([60, 40, 55]))
```

## Problem 4: Cab Fare Estimator Program

```python

def cab_fare(km, peak):

    fare = 50 + (12 * km)

    if peak:

        fare = fare + (fare * 0.25)

    return fare


print("Output:", cab_fare(10, True))
```

## Problem 5: Employee Attendance Eligibility Program

```python

def attendance_check(attendance):

    present = 0

    for day in attendance:

        if day == "P":

            present += 1

    percent = (present / len(attendance)) * 100

    if percent >= 75:

        return percent, "Eligible"

    return percent, "Not Eligible"


print("Output:", attendance_check(["P", "A", "P", "P", "P"]))
```

## Problem 6: Password Strength Checker Program

```python

def password_check(pwd):

    has_digit = False

    has_special = False

    for ch in pwd:

        if ch.isdigit():

            has_digit = True

        if ch in "@#$":

            has_special = True

    if len(pwd) >= 8 and has_digit and has_special:

        return "Strong Password"

    return "Weak Password"


print("Output:", password_check("Pass@123"))
```
