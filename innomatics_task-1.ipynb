## 1. User Login Check

```python
stored_username = "admin"
stored_password = "1234"

username = "admin"
password = "1234"

# Check login validity
if username == stored_username and password == stored_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

## 2. Pass / Fail Analyzer

```python
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

# Analyzing each student's marks
for score in marks:
    if score >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```

## 3. Simple Data Cleaner

```python
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()  # Removing spaces and converting to lowercase
    cleaned_names.append(cleaned_name)

print("\nCleaned Names:")
print(cleaned_names)
```

## 4. Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]

print("\nMessage Length Analysis")

for message in messages:
    length = len(message)
    print(f"Message: '{message}' | Length: {length}")
    
    # Flag long messages
    if length > 10:
        print("  -> Flag: Message is longer than 10 characters")
```

## 5. Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

# Counting ERROR entries
for log in logs:
    if log == "ERROR":
        error_count += 1

print("\nSystem Log Analysis")
print("Total ERROR entries:", error_count)
```
